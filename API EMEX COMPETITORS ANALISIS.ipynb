```python
from collections import OrderedDict

def flatten(json_object, container=None, name=''):
    if container is None:
        container = OrderedDict()
    if isinstance(json_object, dict):
        for key in json_object:
            flatten(json_object[key], container=container, name=name + key + '_')
    elif isinstance(json_object, list):
        for n, item in enumerate(json_object, 1):
            flatten(item, container=container, name=name + str(n) + '_')
    else:
        container[str(name[:-1])] = str(json_object)
    return container
```

```python
def original_data_collector(data_count_limit,response):
    seller=[]
    seller_rating=[]
    good_seller_cases_rate=[]
    bad_seller_cases_rate=[]
    price=[]
    delivery=[]
    refused_count=[]
    quantity=[]
    order_count=[]
    part_num=[]
    manufactor=[]
    name=[]
    x=0
    
                 
    for i in range(0, min(len(response.json()['searchResult']['originals'][0]['offers']),data_count_limit)):
              part_num.append(response.json()['searchResult']['num'])
              manufactor.append(response.json()['searchResult']['make']) 
            
              seller.append(response.json()['searchResult']['originals'][0]['offers'][i]['rating2']['code'])
              seller_rating.append(response.json()['searchResult']['originals'][0]['offers'][i]['rating2']['rating'])
              good_seller_cases_rate.append(response.json()['searchResult']['originals'][0]['offers'][i]['rating2']['goodPercent'])
              x=response.json()['searchResult']['originals'][0]['offers'][i]['rating2']['goodPercent']
              bad_seller_cases_rate.append(response.json()['searchResult']['originals'][0]['offers'][i]['rating2']['badPercent'])
              price.append(response.json()['searchResult']['originals'][0]['offers'][i]['data']['displayPrice']['value'])
              delivery.append(response.json()['searchResult']['originals'][0]['offers'][i]['data']['delivery']['value'])
              quantity.append(response.json()['searchResult']['originals'][0]['offers'][i]['quantity'])
              name.append(response.json()['searchResult']['name'])
              if x==0:
                  order_count.append(0)
                  refused_count.append(0)
              else:
                  order_count.append(response.json()['searchResult']['originals'][0]['offers'][i]['data']['pricelistRating']['orderCount'])
                  refused_count.append(response.json()['searchResult']['originals'][0]['offers'][i]['data']['pricelistRating']['refusedNew'])
    
    data = {'part_num': part_num,'manufactor': manufactor,'seller': seller, 'seller_rating': seller_rating,'order_count': order_count,'refused_count': refused_count, 'good_seller_cases_rate': good_seller_cases_rate,
            'bad_seller_cases_rate': bad_seller_cases_rate,'price': price,'delivery': delivery,'refused_count': refused_count,
           'quantity': quantity, 'name':name
           }
    df = pd.DataFrame(data)
    
    return df


    
```
